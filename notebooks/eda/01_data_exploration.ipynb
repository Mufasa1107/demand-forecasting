# Retail Sales EDA

This notebook explores synthetic hierarchical retail sales data and checks the shape, seasonality, and promo effects.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from src.data.generate_dummy_data import generate_sales_dataset

df = generate_sales_dataset(regions=['North','South'], stores_per_region=2, categories_per_store=2, skus_per_category=2, periods=180, seed=42)
print(df.head())
print(df.shape)
```

```python
weekly = df.groupby(['date','region'])['demand'].sum().reset_index()
sns.lineplot(data=weekly, x='date', y='demand', hue='region')
plt.title('Demand by region over time')
plt.show()
```
